

# b. Bar chart   

```python
data2 = pd.read_csv('data.csv',index_col=10)

data_2 = pd.DataFrame(data2,columns=['dept_1 sales','dept_2 sales','dept_3 sales'])

total_sales = pd.DataFrame([[data_2['dept_1 sales'].sum(),data_2['dept_2 sales'].sum(),data_2['dept_3 sales'].sum()]],columns=["dept_1 sales","dept_2 sales","dept_3 sales"],index=["Total sales of departments"])

total_sales.T.plot.bar()
plt.title('Bar Chart for Total sales of departments')
```

```python
data_3 = pd.DataFrame(data2,columns=['MarkDown1','MarkDown2','MarkDown3','MarkDown4','MarkDown5'])
data_3 = data_3.dropna()

total_markdown = pd.DataFrame([[data_3['MarkDown1'].sum(),data_3['MarkDown2'].sum(),data_3['MarkDown3'].sum(),data_3['MarkDown4'].sum(),data_3['MarkDown5'].sum()]],columns=["MarkDown1","MarkDown2","MarkDown3","MarkDown4","MarkDown5"],index=["Total Values of Markdowns"])

total_markdown.T.plot.bar()
plt.title('Bar chart for Total values of MarkDown1,2,3,4 and 5')
```

```python
sales_holiday=data_2.loc["Holiday"]

total_sales_holiday = pd.DataFrame([[sales_holiday['dept_1 sales'].sum(),sales_holiday['dept_2 sales'].sum(),sales_holiday['dept_3 sales'].sum()]],columns=["dept_1 sales","dept_2 sales","dept_3 sales"],index=["Total sales of departments on holiday"])

sales_regday = data_2.loc["Regular day"]

total_sales_regday = pd.DataFrame([[sales_regday['dept_1 sales'].sum(),sales_regday['dept_2 sales'].sum(),sales_regday['dept_3 sales'].sum()]],columns=["dept_1 sales","dept_2 sales","dept_3 sales"],index=["Total sales of departments on regular day"])


total_sales_break=total_sales_regday.append(total_sales_holiday)

total_sales_break.T.plot.bar()

plt.title('Bar Chart for Total sales of departments on regular day and holiday')
```

```python
avg_sales_holiday = pd.DataFrame([[sales_holiday['dept_1 sales'].mean(),sales_holiday['dept_2 sales'].mean(),sales_holiday['dept_3 sales'].mean()]],columns=["dept_1 avg sales","dept_2 avg sales","dept_3 avg sales"],index=["Avg holiday"])

avg_sales_regday = pd.DataFrame([[sales_regday['dept_1 sales'].mean(),sales_regday['dept_2 sales'].mean(),sales_regday['dept_3 sales'].mean()]],columns=["dept_1 avg sales","dept_2 avg sales","dept_3 avg sales"],index=["Avg regular day"])


avg_sales_break=avg_sales_regday.append(avg_sales_holiday)

avg_sales_break.T.plot.bar()

plt.title('Bar Chart for average sales of departments on regular day and holiday')
```

```python
markdown_holiday=data_3.loc["Holiday"]

total_markdown_holiday = pd.DataFrame([[markdown_holiday['MarkDown1'].sum(),markdown_holiday['MarkDown2'].sum(),markdown_holiday['MarkDown3'].sum(),markdown_holiday['MarkDown4'].sum(),markdown_holiday['MarkDown5'].sum()]],columns=["Total Markdown 1","Total Markdown 2","Total Markdown 3","Total Markdown 4","Total Markdown 5"],index=["Total markdown on holidays"])

markdown_regday = data_3.loc["Regular day"]

total_markdown_regday = pd.DataFrame([[markdown_regday['MarkDown1'].sum(),markdown_regday['MarkDown2'].sum(),markdown_regday['MarkDown3'].sum(),markdown_regday['MarkDown4'].sum(),markdown_regday['MarkDown5'].sum()]],columns=["Total Markdown 1","Total Markdown 2","Total Markdown 3","Total Markdown 4","Total Markdown 5"],index=["Total markdown on regularday"])


total_markdown_break=total_markdown_regday.append(total_markdown_holiday)

total_markdown_break.T.plot.bar()

plt.title('Bar Chart for Total markdown on regular day and holiday')
```

```python
avg_markdown_holiday = pd.DataFrame([[markdown_holiday['MarkDown1'].mean(),markdown_holiday['MarkDown2'].mean(),markdown_holiday['MarkDown3'].mean(),markdown_holiday['MarkDown4'].mean(),markdown_holiday['MarkDown5'].mean()]],columns=["Avg Markdown 1","Avg Markdown 2","Avg Markdown 3","Avg Markdown 4","Avg Markdown 5"],index=["Avg markdown on holidays"])

avg_markdown_regday = pd.DataFrame([[markdown_regday['MarkDown1'].mean(),markdown_regday['MarkDown2'].mean(),markdown_regday['MarkDown3'].mean(),markdown_regday['MarkDown4'].mean(),markdown_regday['MarkDown5'].mean()]],columns=["Avg Markdown 1","Avg Markdown 2","Avg Markdown 3","Avg Markdown 4","Avg Markdown 5"],index=["Avg markdown on regularday"])

avg_markdown_break=avg_markdown_regday.append(avg_markdown_holiday)

avg_markdown_break.T.plot.bar()

plt.title('Bar Chart for average markdown on regular day and holiday')
```
